```python
def _get_namelist():
    '''无需参数直接调用'''
    dataframe = pd.read_excel(r"E:\Ipython\Machine Learning\ml\name.xlsx",encoding_override='gbk')
    df = dataframe.drop(['index'],axis=1)
    gene = [30,40,50,'60pre','60pos',70,80]
    name_list = []
    for i in gene:
        gene_name = np.array(df[df['Generations'] == i ]['Name'])
        name_list.append(gene_name)
    return name_list
```

```python
def find_directory(gene_list,name_list):
    sin_co = ['single-txt','co-txt']
    try:
        i = 0
        sc_paths=[]
        for gene in gene_list:
            dir = 'F:/paper'
            gene_path = join(str(dir+ "download/"),str(gene))
            for j in range(len(name_list[i])):
                name_path = gene_path + "/"+ name_list[i][j]
                for sc in sin_co:
                    sc_path = name_path + "/"+ str(sc)
                    sc_paths.append(sc_path)
            i+=1
        return sc_paths
    except:
        raise ValueError
```

```python
#在人名下创建新的文件夹txt
def make_txt_directory(gene_list,name_list):
    try:
        i = 0
        for gene in gene_list:
            gene_path = r'F:\paperdownload\\'+ str(gene)
            for j in range(len(name_list[i])):
                name_path = gene_path + '\\'+ name_list[i][j]
                sc_path = name_path + '\\'+ str('txt')
                if os.path.exists(sc_path) == 0:
                    os.makedirs(sc_path)
            i += 1
    except:
        raise ValueError
#调用     
name_list = _get_namelist()
gene = [30,40,50,'60pre','60pos',70,80]
make_txt_directory(gene,name_list)
```
